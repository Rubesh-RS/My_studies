```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

print(sns.get_dataset_names())
```

```python
crash_df=sns.load_dataset('car_crashes')
```

```python
crash_df.head()
```

```python
#distribution plot
#>>>>>histogram plot
#kde-Kernal density estimaton
# sns.distplot(crash_df['alcohol'])
sns.displot(crash_df['alcohol'])
```

```python
sns.jointplot(x='not_distracted',y='speeding',data=crash_df,kind='hist')#kind=hist,kde,hex
```

```python
sns.pairplot(crash_df,hue='abbrev')
```

```python
sns.set_style('white')

plt.figure(figsize=(10,8))

sns.set_context('paper', font_scale=1.5)#talk, paper, poster
sns.jointplot(x='speeding', y='alcohol', data = crash_df, kind= 'reg')
# sns.despine(top=False, bottom=False)
```

```python
#CATAGORICAL PLOTS
```

```python
tips_df=sns.load_dataset('tips')
tips_df.head()
```

```python
sns.barplot(x='sex', y='total_bill', data=tips_df, hue= 'day', estimator=np.mean)#we an self define any function and add here in estimator
```

```python
sns.countplot(x='sex', hue='sex' , data = tips_df)
```

```python
tips_df.head()
```

```python
sns.boxplot(x='day', y= 'total_bill', data=tips_df, hue='sex')
plt.legend(loc=0)
```

```python
sns.violinplot(x='day', y= 'total_bill', data=tips_df, hue='sex',split=True)
plt.legend(loc=0)
```

```python
sns.stripplot(x='day', y= 'total_bill', data=tips_df, hue='sex', jitter=True, dodge=True)
plt.legend(loc=0)
plt.figure(figsize=(10,5))
```

```python
sns.violinplot(x='day', y= 'total_bill', data=tips_df, hue='sex', split=True)
sns.swarmplot(x='day', y= 'total_bill', data=tips_df, hue='sex',color='white')
plt.legend(loc=0)
```

```python
sns.set_style('dark')
sns.set_context('paper')
sns.stripplot(x='day', y= 'total_bill', data=tips_df, hue='sex',palette='magma')
plt.legend(loc=0)
plt.figure(figsize=(10,5))
```

```python
flight_df=sns.load_dataset('flights')
flight_df.head()
```
