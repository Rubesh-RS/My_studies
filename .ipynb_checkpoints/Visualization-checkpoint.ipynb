```python
import numpy as np
import matplotlib.pyplot as plt

# Plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 2, 1)  # Change to a 2x2 grid
plt.plot(x, y)

# Plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 2, 2)
plt.plot(x, y)

# Plot 3:
x = np.array([0, 1, 2, 3])
y = np.array([15, 25, 35, 45])

plt.subplot(2, 2, 3)
plt.plot(x, y)

# Plot 4:
x = np.array([0, 1, 2, 3])
y = np.array([5, 15, 25, 35])

plt.subplot(2, 2, 4)
plt.plot(x, y)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
z = np.array([2,1,4,1])

fig, ax = plt.subplots(3, 1)  # Change to a 2x2 grid

ax[0].plot(x,y,marker = "o", color = "r", linestyle = "--", label = "xy")

ax[1].plot(y,z,marker = "o", color = "g", linestyle = "--", label = "yz")

ax[2].plot(x,z,marker = "o", color = "b", linestyle = "--", label = "xz")

ax[0].set_xlabel("TimeX")
ax[0].set_ylabel("TimeY")
ax[0].legend()
ax[1].set_xlabel("MoneyX")
ax[1].set_ylabel("MoneyY")
ax[1].legend()
ax[2].set_xlabel("HealthX")
ax[2].set_ylabel("HealthY")
ax[2].legend()

plt.tight_layout()
plt.show()
```
