```python
import numpy as np
import matplotlib.pyplot as plt

# Plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
z = np.array([2,1,4,1])

fig, ax = plt.subplots(3, 1)  # Change to a 2x2 grid

ax[0].plot(x,y,marker = "o", color = "r", linestyle = "--", label = "xy")

ax[1].plot(y,z,marker = "o", color = "g", linestyle = "--", label = "yz")

ax[2].plot(x,z,marker = "o", color = "b", linestyle = "--", label = "xz")

ax[0].set_xlabel("TimeX")
ax[0].set_ylabel("TimeY")
ax[0].legend()
ax[1].set_xlabel("MoneyX")
ax[1].set_ylabel("MoneyY")
ax[1].legend()
ax[2].set_xlabel("HealthX")
ax[2].set_ylabel("HealthY")
ax[2].legend()
fig.suptitle("Test3")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
z = np.array([2, 1, 4, 1])

fig, ax = plt.subplots(3, 3, figsize=(10, 10))  # Create a 3x3 grid of subplots

# Plot 1: Top-left
ax[0, 0].plot(x, y, marker="o", color="r", linestyle="--", label="xy")
ax[0, 0].set_xlabel("TimeX")
ax[0, 0].set_ylabel("TimeY")
ax[0, 0].legend()

# Plot 2: Top-middle
ax[0, 1].plot(y, z, marker="o", color="g", linestyle="--", label="yz")
ax[0, 1].set_xlabel("MoneyX")
ax[0, 1].set_ylabel("MoneyY")
ax[0, 1].legend()

# Plot 3: Top-right
ax[0, 2].plot(x, z, marker="o", color="b", linestyle="--", label="xz")
ax[0, 2].set_xlabel("HealthX")
ax[0, 2].set_ylabel("HealthY")
ax[0, 2].legend()

# Plot 4: Middle-left
ax[1, 0].plot(x, y**2, marker="s", color="m", linestyle="-.", label="xy^2")
ax[1, 0].set_xlabel("X^2 Time")
ax[1, 0].set_ylabel("Squared Y")
ax[1, 0].legend()

# Plot 5: Middle-center
ax[1, 1].plot(z, x, marker="^", color="c", linestyle=":", label="zx")
ax[1, 1].set_xlabel("HealthZ")
ax[1, 1].set_ylabel("TimeX")
ax[1, 1].legend()

# Plot 6: Middle-right
ax[1, 2].plot(y, z**2, marker="x", color="y", linestyle="--", label="y*z^2")
ax[1, 2].set_xlabel("MoneyX")
ax[1, 2].set_ylabel("Squared HealthY")
ax[1, 2].legend()

# Plot 7: Bottom-left
ax[2, 0].plot(y, x**2, marker="d", color="orange", linestyle="-", label="yx^2")
ax[2, 0].set_xlabel("MoneyY")
ax[2, 0].set_ylabel("Squared TimeX")
ax[2, 0].legend()

# Plot 8: Bottom-middle
ax[2, 1].plot(x, z**3, marker="p", color="purple", linestyle="-.", label="xz^3")
ax[2, 1].set_xlabel("TimeX")
ax[2, 1].set_ylabel("Cubed HealthZ")
ax[2, 1].set_ylabel("Cubed HealthZ")
ax[2, 1].legend()

# Plot 9: Bottom-right
ax[2, 2].plot(z, y, marker="*", color="brown", linestyle=":", label="zy")
ax[2, 2].set_xlabel("HealthZ")
ax[2, 2].set_ylabel("MoneyY")
ax[2, 2].legend()

# Add a global title
fig.suptitle("3x3 Subplots Example", fontsize=16)

plt.tight_layout()
plt.show()
```
