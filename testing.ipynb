```python
import pandas as pd

# Sample DataFrame
data = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df = pd.DataFrame(data)

# Create a new column (e.g., 'C') with some values
df['C'] = [7, 8, 9]

# Reorder columns to place 'C' as the second column
cols = df.columns.tolist()
cols.insert(1, cols.pop(cols.index('C')))  # Move 'C' to the second position
df = df[cols]
print(df)
```

```python
df
```

```python
new_col = pd.DataFrame([["Quathar","North","Rubesh","7/30/2014 0:00:00","200000","1000"]],columns=df.columns.values)
```

```python
new_col
```

```python
df = pd.concat([df,new_col],ignore_index=True)
```

```python
df.sort_values(["Country","Region"], inplace=True)
```
